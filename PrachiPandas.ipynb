```python
df4= df3.dropna(axis = 1)        #remove all columns which atleast contain 1 null
df4
```

```python
df4.shape
```

```python
a=df3.dropna(how= 'any')            #remove row if any 1 is null
```

```python
b=df3.dropna(how='all')
```
