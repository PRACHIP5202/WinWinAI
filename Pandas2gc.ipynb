### **Have a look at num 6, Computer for better understanding**

forword fill - prev value

backword fill - nxt value

```python
df.fillna({'Physics': 99, 'Chemistry':89, 'Maths':100, 'Computer':100})
```

```python
df.fillna(method = 'ffill')                 # Forward fill, filled with the closest upper value in the col, || if axis=1 then fill with pre value in row
```

```python
df.fillna(method= 'ffill', axis=1)                     # forword fill but prev value in row
```

```python
df.fillna(method = 'bfill')                     # Backword fill,  fill with next value in a col
```

```python
df.fillna(method = 'bfill', axis=1)                  #fill with nxt value in a row, seee if nt existes then NaN i.e. Null
```
