## Merge()

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Physics':[99,98,97,9,6]})
df1
```

```python
df2= pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Chemistry': [88,89,67,78,63]})
df2
```

```python
pd.merge(df1,df2, on='Roll No.')
```

```python
pd.merge(df2,df1, on='Roll No.')
```

```python
pd.merge(df1,df2)                      #by default on is taken as the common row in both
```

```python
df3= pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Physics': [33,4,32,32,44]})
df4= pd.DataFrame({'Roll No.':[1,4,5,7,8], 'Chemistry': [45,545,65,32,23]})
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4, how='left')
```

```python
pd.merge(df3,df4, how='right')
```

```python
pd.merge(df3,df4, how='outer')
```

append func is removed now we have concat

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Chemistry':[43,5,65,2,8], 'Computer':[23,45,656,87,4]})
df2= pd.DataFrame({'Roll No.':[6,7,8,9,10], 'Physics':[43,5,65,2,8], 'Computer':[23,45,656,87,4]})
print(df1)
print(df2)
```

```python
pd.concat([df1,df2])                           #concatinate successfully but look at the index, its nt in order
```

```python
pd.concat([df1,df2], ignore_index=True)                                 #index is ignord n set as per arrangement needed
```

```python
pd.concat([df2,df1], ignore_index=True)                 #df2 then df1
```

```python
pd.concat([df1,df2], ignore_index=True, sort=True)          #sorts the column name
```
